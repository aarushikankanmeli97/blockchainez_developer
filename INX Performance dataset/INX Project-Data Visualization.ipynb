## Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

## Data Visualization

```python
#Readjusting the size of the plot
rcParams['figure.figsize'] = 15,10
```

```python
#"crosstab" is used to visualize performance rating in one plot
df = pd.crosstab(index = data['EmpDepartment'],columns = data['PerformanceRating'])
df.columns = ["Good","Better","Best"]
df
```

```python
df.plot(kind = "bar")
plt.xlabel('EmpDepartment',fontsize = 14)
plt.ylabel('Number of Employees',fontsize = 14)
plt.title('Department wise performance',fontsize = 16)
plt.xticks(fontsize = 14,rotation = 15)
plt.savefig('DEpartment wise performance.png')
```

```python
# Using correlation to find most affected factors in employee performance
new_df = data.corr()
```

```python
# Sorting the values to determine the three most affected factors
new_df.sort_values(by = 'PerformanceRating', inplace=True, ascending=False)
new_df[['PerformanceRating']]
```

```python
# Bar plot to check the performance rating with respect to each factor
# Plot for "Employee work life balance"
plot_df = pd.crosstab(index = data['EmpWorkLifeBalance'],columns = data['PerformanceRating'])
plot_df.columns = ["Good","Better","Best"]
plot_df.reset_index(inplace=True)
plot_df
```

```python
x = np.arange(4)
lab = ['Low', 'Medium', 'High', 'Very High']
plt.xticks(x,lab,rotation = 15, fontsize = 14)
plt.bar(x, plot_df.Good, width = 0.1, label="Good")
plt.bar(x+0.1, plot_df.Better, width = 0.1, label="Better")
plt.bar(x+0.2, plot_df.Best, width = 0.1, label="Best")
plt.xlabel('Employee Work Life Balance',fontsize = 14)
plt.ylabel('Number of Employee',fontsize = 14)
plt.title('Work life Balance wise performance',fontsize = 16)
plt.legend(shadow=True)
plt.savefig('Work life balance wise performance')
```

```python
#Plot for "Employee last salary hike percent"
plot_df1 = pd.crosstab(index = data['EmpLastSalaryHikePercent'],columns = data['PerformanceRating'])
plot_df1.columns = ["Good","Better","Best"]
plot_df1.reset_index(inplace=True)
plot_df1
```

```python
x = np.arange(15)
plt.xticks(x, plot_df1.EmpLastSalaryHikePercent,rotation = 15,fontsize = 14)
plt.bar(x,plot_df1.Good,width = 0.3,label = "Good")
plt.bar(x + 0.2,plot_df1.Better,width = 0.3,label = "Better")
plt.bar(x + 0.4,plot_df1.Best,width = 0.3,label = "Best")
plt.xlabel('Employee last salary hike percentage',fontsize = 14)
plt.ylabel('NUmber of Employee',fontsize = 14)
plt.title('Salary hike wise Performance',fontsize = 16)
plt.legend(shadow = True)
plt.savefig('Salary hike wise performance')
```

```python
# Plot for "Employee environment satisfaction"
plot_df2 = pd.crosstab(index = data['EmpEnvironmentSatisfaction'],columns = data['PerformanceRating'])
plot_df2.columns = ["Good","Better","Best"]
plot_df2.reset_index(inplace = True)
plot_df2
```
